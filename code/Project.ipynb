```r
emp.rep=subset(emp, Govenor=='Republican')
emp.dem=subset(emp, Govenor=='Democrat')
```

```r
plot(density(emp.rep$MurderRate, bw="SJ"), main='Murder Rates when Republican is in office', xlim=c(10,100), ylim=c(0,.04))

plot(density(emp.dem$MurderRate, bw="SJ"), main='Murder Rates when Democrats is in office', xlim=c(10,100), ylim=c(0,.04))

```

```r
library('sm')
rep=bw.SJ(emp$MurderRate[emp$Govenor=='Republican'])
dem=bw.SJ(emp$MurderRate[emp$Govenor=='Democrat'])
h=2*(1/rep+1/dem)^-1
sm.density.compare(emp$MurderRate, group=emp$Govenor, model="equal", bw=h, nboot=10000, xlim=c(10,100), ylim=c(0,.04))



```
