```r
library('sm')
```

```r
rep=bw.SJ(emp$MurderRate[emp$Govenor=='Republican'])
dem=bw.SJ(emp$MurderRate[emp$Govenor=='Democrat'])
h=2*(1/rep+1/dem)^-1
sm.density.compare(emp$MurderRate, group=emp$Govenor, model="equal", bw=h, nboot=10000, xlim=c(10,100), ylim=c(0,.04))
```

```r
#rep
#dem
mean_rep=mean(emp$MurderRate[emp$Govenor=='Republican'])
mean_dem=mean(emp$MurderRate[emp$Govenor=='Democrat'])

sd_rep=sd(emp$MurderRate[emp$Govenor=='Republican'])
sd_dem=sd(emp$MurderRate[emp$Govenor=='Democrat'])

emp_std=c((emp$MurderRate[emp$Govenor=='Republican']-mean_rep)/sd_rep,
          +(emp$MurderRate[emp$Govenor=='Democrat']-mean_dem)/sd_dem)



r=bw.SJ(emp_std[emp$Govenor=='Republican'])
d=bw.SJ(emp_std[emp$Govenor=='Democrat'])


h=2*(1/r+1/d)^-1



sm.density.compare(emp_std, group=emp$Govenor, model="equal", bw=h, nboot=10000, xlim=c(10,100), ylim=c(0,.04))


```

```r
#semi nonparametric regression
library('np')
trawl.bw=npplregbw(emp$MurderRate~emp$Month|emp$Unemployment.Rate, regtype="ll")
trawl.res=npplreg(trawl.bw)
trawl.plot=npplreg(trawl.bw, exdat=seq(3.3,11.4, .01), ezdat=seq(3.3,11.4,.01))$mean
plot(seq(3.3,11.4,.01), trawl.plot, type="l", xlab="Unemployment Rate", ylab="estimate of m")
```
